```python
import json

with open("record.json") as json_file:
    data = json.load(json_file)

name = input("\nPlease Enter Your Name : ")
print(f"\nHello {name},")
print("WELCOME TO 'BREAD & BUTTER' ONLINE SHOPPING PORTAL")
print("--------------------------------------------------")
print("'BREAD & BUTTER' is a family-owned organization with the enviable status while in the dessert manufacturing marketplace for its premium cakes distinctiveness.", end = " ")
print("We provide a variety of wholesale items including Breads, Cakes, Cold-Drinks, Chocolate, Biscuits, Floavours, Pizza Base, Sauces, Pastries and a lot more.")


print("\nAlso we have a special offer for you : ")
print("----------------------------------------")
print("On shopping of 5000 or above ---  10 % discount(max discount of 1000)")

print("\nBelow are the list of items. You can now choose your items....")

print("|-----------|----------------|---------------------|---------------|-------------|------------------|--------------|")
print("|    ID     |     Category   |   Name of item      |     Weight    |     Cost    |       Rating     |   Quantity   |")
print("|           |                |                     |               |  (per unit) |   (by customers) |  (in stock)  |")
print("|-----------|----------------|---------------------|---------------|-------------|------------------|--------------|")

for i in data:
    print('|', i, "   |\t", data[i]["Category"], "|\t", data[i]["Name"], "   |\t", data[i]["Weight"], " |\t", data[i]["Cost"], "    |\t", data[i]["Rating"],"       |\t", data[i]["Qty"], "     |")
    print("|-----------|----------------|---------------------|---------------|-------------|------------------|--------------|")



n = int(input("\nEnter number of items you want to buy: "))
itemID  = []
itemQty = []
for i in range(n):
    x = input(f"Enter ID of {i+1} item: ")
    y = input(f"Enter the quantity of item with ID {x} you want to buy: ")
    itemID.append(x)
    itemQty.append(y)


print("\nHere are the items that you selected...")

print("|-----------|----------------|---------------------|---------------|-------------|--------------|")
print("|    ID     |     Category   |   Name of item      |     Weight    |     Cost    |   Quantity   |")
print("|           |                |                     |               |  (per unit) |              |")
print("|-----------|----------------|---------------------|---------------|-------------|--------------|")
j = 0
for i in itemID:
    print('| ', i, "  |\t", data[i]["Category"], "|\t", data[i]["Name"], "   |\t", data[i]["Weight"], " |\t", data[i]["Cost"],"    |\t", itemQty[j], "    |")
    print("|-----------|----------------|---------------------|---------------|-------------|--------------|")
    j = j + 1
    
   
confirm = input("\nPress Y to confirm your items....")
j = 0
k = 0

if(confirm == "Y"):
    itemPrice = []
    for i in itemID:
        x = int(data[i]["Cost"]) * int(itemQty[j])
        itemPrice.append(int(x))
        j = j + 1
        
    print("\nPlease do checkout and pay the amount.......")
    print("|-----------|----------------|---------------------|---------------|-------------|--------------|-------------|")
    print("|    ID     |     Category   |   Name of item      |     Weight    |     Cost    |   Quantity   |   Price     |")
    print("|           |                |                     |               |  (per unit) |              |             |")
    print("|-----------|----------------|---------------------|---------------|-------------|--------------|-------------|")
    for i in itemID:
        print('| ', i, "  |\t", data[i]["Category"], "|\t", data[i]["Name"], "   |\t", data[i]["Weight"], " |    ", data[i]["Cost"],"    |\t", itemQty[k], "    |   ", itemPrice[k], "    |")
        print("|-----------|----------------|---------------------|---------------|-------------|--------------|-------------|")
        k = k + 1
        
    bill = 0
    for i in itemPrice:
        bill = bill + i
        
    
    if(bill >= 5000 and bill <= 10000):
        pay = bill - 0.10 * (bill)
        print("|                                                                                Total Value    |   ", bill,"    |")
        print("|-----------------------------------------------------------------------------------------------|-------------|")
        print("|                                                                               Discount(10 %)  |   ", 0.1 * bill,"   |")
        print("|-----------------------------------------------------------------------------------------------|-------------|")
        print("|                                                                            Amount To Be Paid  |   ", pay,"  |")
    elif(bill < 5000):
        pay = bill - 0.0
        print("|                                                                                Total Value    |   ", bill,"    |")
        print("|-----------------------------------------------------------------------------------------------|-------------|")
        print("|                                                                       Discount(Not Eligible)  |   ", 0.0,"     |")
        print("|-----------------------------------------------------------------------------------------------|-------------|")
        print("|                                                                            Amount To Be Paid  |   ", pay,"  |")
    elif(bill > 10000):
        pay = bill - 1000.0
        print("|                                                                                Total Value    |   ", bill,"   |")
        print("|-----------------------------------------------------------------------------------------------|-------------|")
        print("|                                                                                     Discount  |   ", 1000.0,"  |")
        print("|-----------------------------------------------------------------------------------------------|-------------|")
        print("|                                                                            Amount To Be Paid  |   ", pay," |")
    print("|-----------------------------------------------------------------------------------------------|-------------|")
    
    from datetime import timedelta, date, datetime
    import time, uuid

    today = date.today()

    t = time.localtime()
    timeNow = time.strftime("%H:%M:%S", t)
    
    newdate = today + timedelta(days=1)
    
    trans = uuid.uuid1()
    
    paid = input("\nPress Y if you have paid the amount : ")
        
    if(paid == "Y"):   
         
        address = input("Enter your shipping address : ")
        
        print("\nThank you for shopping at 'BREAD & BUTTER'. Find your order receipt below....")
        print("\n")
        print("---------------------|------------------------------------------------------------------------------------")
        print(f"   Name              |    {name}")
        print("---------------------|------------------------------------------------------------------------------------")
        print(f"   Transaction ID    |    {trans}")
        print("---------------------|------------------------------------------------------------------------------------")
        print(f"   Date              |    {today}")
        print("---------------------|------------------------------------------------------------------------------------")
        print(f"   Time              |    {timeNow}")
        print("---------------------|------------------------------------------------------------------------------------")
        print(f"   Amount Paid       |    {pay}")
        print("---------------------|------------------------------------------------------------------------------------")
        print(f"   Shipping Address  |    {address}")
        print("---------------------|------------------------------------------------------------------------------------")
        print(f"   Deliever Date     |    {newdate}")
        print("---------------------|------------------------------------------------------------------------------------")
        
            
else:
    print("Choose the items again.")
    

with open("record.json") as json_file:
    data = json.load(json_file)

k = 0
for i in data:
    for j in itemID:
        if(i==j):
            data[i]["Qty"] = str(int(data[i]["Qty"]) - int(itemQty[k])) + " "
            k = k + 1
        
fd = open('record.json', 'w')
json.dump(data, fd)
fd.close()

detail = {str(trans):{"Name" : name, "Transaction ID" : str(trans), "Date" : str(today), "Time" : timeNow, "Amount Paid" : pay,
                "Shipping Address" : address, "Deliever Date": str(newdate)}}

fs = open('sale detail.json', 'a')
fs.write('\n\n')
json.dump(detail, fs)
fs.close()
            
```
