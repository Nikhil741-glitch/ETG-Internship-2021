```python
# importing libraries
import json, time, uuid
from datetime import timedelta, date, datetime


# opening and reading the record.json file
fd = open('record.json', 'r')
d = fd.read()
fd.close()


# taking the record of items in dictionary form into data variable using loads function 
data = json.loads(d)
    
    
# taking user's name
name = input("\nPlease Enter Your Name : ")


# greetings from the company
print(f"\nHELLO {name},\n")
print("\t\t\t\t========================================================")
print("\t\t\t\t   WELCOME TO 'BREAD & BUTTER' ONLINE SHOPPING PORTAL   ")
print("\t\t\t\t========================================================")
print('\n')


# printing intro of the company
print("===============================================================================================================================")
print("'BREAD & BUTTER' is a family-owned organization with the enviable status while in the dessert manufacturing marketplace for its premium cakes distinctiveness.", end = " ")
print("We provide a variety of wholesale items including Breads, Cakes, Cold-Drinks, Chocolate, Biscuits, Floavours, Pizza Base, Sauces, Pastries and a lot more.")
print("===============================================================================================================================")
print('\n')


# printing discount offer
print("===============================================================================================================================")
print("            SPECIAL OFFER FOR YOU --------> SHOP FOR 5000 OR ABOVE AND GET 10 % DISCOUNT (MAX DISCOUNT OF 5000)                ")
print("===============================================================================================================================")
print("\n")


# printing the list of items in the inventory so that the user can select items
print("====================================================================================================================")
print("                   BELOW ARE LISTED VARIOUS ITEMS IN THE SHOP, YOU CAN SELECT YOUR ITEMS NOW")
print("====================================================================================================================")
print("|    ID     |     Category   |   Name of item      |     Weight    |   Cost(Rs)  |       Rating     |   Quantity   |")
print("|           |                |                     |               |  (per unit) |   (by customers) |  (in stock)  |")
print("|-----------|----------------|---------------------|---------------|-------------|------------------|--------------|")
for i in data:
    print(f"|   {i}  |   {data[i]['Category']}  |    {data[i]['Name']}   |     {data[i]['Weight']}  |\t {data[i]['Cost']}\t |\t   {data[i]['Rating']}\t    |\t   {data[i]['Qty']}     |")
    print("|-----------|----------------|---------------------|---------------|-------------|------------------|--------------|")

    
# taking input how many items user want to buy
n = int(input("\nEnter number of items you want to buy: "))
itemSelected  = {}
for i in range(n):
    # taking id of the item
    x = input(f"Enter ID of {i+1} item: ")
    # taking quantity of the item
    y = int(input(f"Enter the quantity of item with ID {x} you want to buy: "))
    itemSelected[x] = y
print('\n')  


# printing the selected items by the user for review
print("=================================================================================================")
print("                                  PLEASE RECHECK YOUR ITEMS")
print("=================================================================================================")
print("|    ID     |     Category   |   Name of item      |     Weight    |   Cost(Rs)  |   Quantity   |")
print("|           |                |                     |               |  (per unit) |              |")
print("|-----------|----------------|---------------------|---------------|-------------|--------------|")
for i in itemSelected:
    print(f'|   {i}  |\t{data[i]["Category"]}  |\t  {data[i]["Name"]}   |\t {data[i]["Weight"]}  |\t {data[i]["Cost"]}     |\t{itemSelected[i]}      |')
    print("|-----------|----------------|---------------------|---------------|-------------|--------------|") 
    
    
# taking confirmation message from the user 
confirm = input("\nPress Y to confirm your items....")
print('\n')  


# calculating the price of each item
if(confirm == "Y"):
    itemPrice = {}
    for i in itemSelected:
        x = data[i]["Cost"] * itemSelected[i]
        itemPrice[i] = x;
        
        
    # generating the bill  
    print("===============================================================================================================")
    print("                                    DO CHECKOUT AND MAKE THE PAYMENT")
    print("===============================================================================================================")
    print("|    ID     |     Category   |   Name of item      |     Weight    |     Cost    |   Quantity   |   Price     |")
    print("|           |                |                     |               |  (per unit) |              |             |")
    print("|-----------|----------------|---------------------|---------------|-------------|--------------|-------------|")
    for i in itemSelected:
        print('| ', i, "  |\t", data[i]["Category"], "|\t", data[i]["Name"], "   |\t", data[i]["Weight"], " |    ", data[i]["Cost"],"    |\t", itemSelected[i], "    |   ", itemPrice[i], "    |")
        print("|-----------|----------------|---------------------|---------------|-------------|--------------|-------------|")
        
        
    # sum up price of all the items
    bill = 0
    for i in itemPrice:
        bill = bill + itemPrice[i]
        
        
    # caluculate billing amount after discount
    if(bill >= 5000 and bill <= 50000):
        discount = int(0.1 * bill)
        pay = bill - discount
        print("|                                                                                Total Value    |   ", bill,"   |")
        print("|-----------------------------------------------------------------------------------------------|-------------|")
        print("|                                                                               Discount(10 %)  |   ", discount,"    |")
        print("|-----------------------------------------------------------------------------------------------|-------------|")
        print("|                                                                            Amount To Be Paid  |   ", pay,"   |")
    elif(bill < 5000):
        discount = 0
        pay = bill - discount
        print("|                                                                                Total Value    |   ", bill,"    |")
        print("|-----------------------------------------------------------------------------------------------|-------------|")
        print("|                                                                       Discount(Not Eligible)  |   ", discount,"       |")
        print("|-----------------------------------------------------------------------------------------------|-------------|")
        print("|                                                                            Amount To Be Paid  |   ", pay,"    |")
    elif(bill > 50000):
        discount = 5000
        pay = bill - discount
        print("|                                                                                Total Value    |   ", bill,"   |")
        print("|-----------------------------------------------------------------------------------------------|-------------|")
        print("|                                                                                     Discount  |   ", 1000.0,"  |")
        print("|-----------------------------------------------------------------------------------------------|-------------|")
        print("|                                                                            Amount To Be Paid  |   ", pay," |")
    print("|-----------------------------------------------------------------------------------------------|-------------|")
    
    
    # generating current date
    today = date.today()
    
    
    # generating current time
    t = time.localtime()
    timeNow = time.strftime("%H:%M:%S", t)
    
    
    # generating delievery date
    delieverDate = today + timedelta(days=2)
    
    
    # generating transaction id
    trans = uuid.uuid1()
    
    
    # taking payment confirmation message
    paid = input("\nPress Y if you have done with the payment : ")
        
        
    if(paid == "Y"):   
        
        # taking the address of the user
        address = input("Enter your shipping address : ")
        print('\n')
        
        # printing the order reciept of user
        print('==========================================================================================================')
        print("             THANK YOU FOR ORDER AT 'BREAD & BUTTER'. PLEASE FIND YOUR ORDER RECIEPT BELOW")
        print('==========================================================================================================')
        print("\n")
        print(f"---------------------|------------------------------------------------------------------------------------")
        print(f"   Name              |    {name}")
        print(f"---------------------|------------------------------------------------------------------------------------")
        print(f"   Transaction ID    |    {trans}")
        print(f"---------------------|------------------------------------------------------------------------------------")
        print(f"   Date              |    {today}")
        print(f"---------------------|------------------------------------------------------------------------------------")
        print(f"   Time              |    {timeNow}")
        print(f"---------------------|------------------------------------------------------------------------------------")
        print(f"   Amount Paid       |    {pay}")
        print(f"---------------------|------------------------------------------------------------------------------------")
        print(f"   Shipping Address  |    {address}")
        print(f"---------------------|------------------------------------------------------------------------------------")
        print(f"   Deliever Date     |    {delieverDate}")
        print(f"---------------------|------------------------------------------------------------------------------------")
        
        
        # updating the record file by subtracting the quantity that user buy
        for i in data:
            if(i in itemSelected):
                data[i]["Qty"] = data[i]["Qty"] - itemSelected[i]
            
            
        # converting dictionary to a string using dumps function
        n = json.dumps(data)
        
        
        # opening the file in write mode and write the updated file 
        fd = open('record.json', 'w')
        fd.write(n)
        fd.close()

        
        # updating sales file 
        detail = {"Name" : name, "Transaction ID" : str(trans), "Date" : str(today), "Time" : str(timeNow), "Amount Paid" : pay,
                "Shipping Address" : address, "Deliever Date": str(delieverDate)}
        
        fs = open('sale detail.json', 'r')
        m = fs.read()
        fs.close()
        
        saleid = str(today) + " " + str(timeNow)
        
        value = json.loads(m)
        value[saleid] = detail
        
        p = json.dumps(value)
        
        # writing the bill into the sale detail.json file 
        fm = open('sale detail.json', 'w')
        fm.write(p)
        fm.close()
        
        
    else:
        print("You clicked something else")
            
else:
    print("You clicked something else please choose the items again.")
 
```
