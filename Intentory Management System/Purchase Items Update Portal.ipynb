```python
# importing library
import json


# opening record.json file in read mode
fs = open('record.json', 'r')
m = fs.read()
fs.close()
     
    
# converting the data into ditionary using loads function
value = json.loads(m)


# taking input whether user want to add items or update the existing items
pur = input("If you want to update new items in the inventory PRESS A , otherwise if you want to update existing item quantity PRESS B: ")

print("\n")
# if user want to add new items
if(pur == "A"):

    # taking input the item details
    n = int(input("Enter how many items you want to add : "))
    for i in range(n):
        ID = input(f"Enter the id of {i+1} item : ")
        category = input("Enter the category of the item : ")
        name = input("Enter the name of the item : ")
        weight = input("Enter the weight of the item : ")
        cost = int(input("Enter the cost of the item : "))
        rating = float(input("Enter the rating of the item by customers who buy it: "))
        quantity = int(input("Enter the quantity of the item : "))
    
        detail = {"Category" : category, "Name" : name, "Weight" : weight, "Cost" : cost, "Rating" : rating, "Qty" : quantity}
        
        # updating item in our file
        value[ID] = detail
        
        print("Successfully added....")
        print("\n")
    
    
# if user want to update the quantity of existing items
elif(pur == "B"):

    n = int(input("Enter how many items quantity you want to update : "))

    for i in range(n):
        
        # taking input the quantity of items
        ID = input(f"Enter the id of {i+1} item : ")
        quantity = int(input("Enter the quantity of the item : "))
        
        # updating the quantity in the file
        value[ID]["Qty"] = value[ID]["Qty"] + quantity
        
        print("Successfully updated....")
        print("\n")

    
p = json.dumps(value)

# writing the updating data into the record.json file
fm = open('record.json', 'w')
fm.write(p)
fm.close()

```
